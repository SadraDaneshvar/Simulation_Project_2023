The code block below defines two functions: `exponential_generator` and `triangular_generator`. Let's break down each function and its purpose:

- `exponential_generator(mean)`: This function generates a random number from an exponential distribution with a specified mean (`mean`) using its inverse cumulative density function.

- `triangular_generator(Min, Max, Mode)`: This function generates a random number from a triangular distribution with specified minimum (`Min`), maximum (`Max`), and mode (`Mode`) values using its inverse cumulative density function.
<span>

```python
def exponential_generator(mean):

    # Generate a random number between 0 and 1
    random_number = random.random()
    
    # Calculate the exponential number
    exp_number = -mean * np.log(random_number)
    
    return exp_number

def triangular_generator(Min, Max, Mode):
    
    # Generate a random number between 0 and 1
    random_number = random.random()
    
    # Calculate the triangular number
    if (random_number >= 0) and (random_number < (Mode-Min)/(Max-Min)):
        tri_number = np.sqrt((Max-Min)*(Mode-Min)*random_number)+Min
    else:
        tri_number = Max - np.sqrt((1-random_number)*(Max-Min)*(Max-Mode))
        
    return tri_number
```

```python
def event_generator_2(env, car_id, mean_photography,
                  min_documentation, max_documentation, mode_documentation,
                  mean_evaluation, min_fulfillment, max_fulfillment,
                  mode_fulfillment, photographer, expert, evaluation_expert):
    
    # Access global variables documentation_queue and fulfillment_queue
    global documentation_queue
    global fulfillment_queue

    # Record the current simulation time as the time the car entered the photography queue
    time_entered_queue_for_photography = env.now

    # Store the time the car entered the photography queue in the list_of_cars
    list_of_cars[car_id][2] = time_entered_queue_for_photography
     
    # Request access to the photographer resource and yield the request
    with photographer.request() as req:
        yield req
    
        # Record the current simulation time as the time the car leaves the photography queue
        time_left_queue_for_photography = env.now

        # Store the time the car left the photography queue in the list_of_cars
        list_of_cars[car_id][3] = time_left_queue_for_photography

        # Calculate the time the car spent waiting in the photography queue
        time_in_queue_for_photography = time_left_queue_for_photography - time_entered_queue_for_photography

        # Store the time spent in the photography queue in the list_of_cars
        list_of_cars[car_id][4] = time_in_queue_for_photography

        # Generate a random sample for the photography time based on an exponential distribution with mean_photography
        sampled_photography_time = exponential_generator(mean_photography)

        # Store the sampled photography time in the list_of_cars
        list_of_cars[car_id][5] = sampled_photography_time

        # Simulate the time it takes for the car's photography to complete using env.timeout
        yield env.timeout(sampled_photography_time)
    
    # Record the current simulation time as the time the car entered the documentation queue
    time_entered_queue_for_documentation = env.now

    # Store the time the car entered the documentation queue in the list_of_cars
    list_of_cars[car_id][6] = time_entered_queue_for_documentation

    # Request access to the documentation expert resource and yield the request
    with expert.request() as req:

        # Increment the documentation queue counter as the car enters the queue
        documentation_queue += 1

        yield req
    
        # Decrement the documentation queue counter as the car leaves the queue
        documentation_queue -= 1

        # Record the current simulation time as the time the car leaves the documentation queue
        time_left_queue_for_documentation = env.now

        # Store the time the car left the documentation queue in the list_of_cars
        list_of_cars[car_id][7] = time_left_queue_for_documentation

        # Calculate the time the car spent waiting in the documentation queue
        time_in_queue_for_documentation = time_left_queue_for_documentation - time_entered_queue_for_documentation

        # Store the time spent in the documentation queue in the list_of_cars
        list_of_cars[car_id][8] = time_in_queue_for_documentation

        # Generate a random sample for the documentation time based on a triangular distribution
        sampled_documentation_time = triangular_generator(min_documentation, max_documentation, mode_documentation)

        # Store the sampled documentation time in the list_of_cars
        list_of_cars[car_id][9] = sampled_documentation_time

        # Simulate the time it takes for the car's documentation to complete using env.timeout
        yield env.timeout(sampled_documentation_time)
    
    # Record the current simulation time as the time the car entered the evaluation queue
    time_entered_queue_for_evaluation = env.now

    # Store the time the car entered the evaluation queue in the list_of_cars
    list_of_cars[car_id][10] = time_entered_queue_for_evaluation

    # Request access to the evaluation expert resource and yield the request
    with evaluation_expert.request() as req:
        yield req
    
        # Record the current simulation time as the time the car leaves the evaluation queue
        time_left_queue_for_evaluation = env.now

        # Store the time the car left the evaluation queue in the list_of_cars
        list_of_cars[car_id][11] = time_left_queue_for_evaluation

        # Calculate the time the car spent waiting in the evaluation queue
        time_in_queue_for_evaluation = time_left_queue_for_evaluation - time_entered_queue_for_evaluation

        # Store the time spent in the evaluation queue in the list_of_cars
        list_of_cars[car_id][12] = time_in_queue_for_evaluation

        # Generate a random sample for the evaluation time based on an exponential distribution with mean_evaluation
        sampled_evaluation_time = exponential_generator(mean_evaluation)

        # Store the sampled evaluation time in the list_of_cars
        list_of_cars[car_id][13] = sampled_evaluation_time

        # Simulate the time it takes for the car's evaluation to complete using env.timeout
        yield env.timeout(sampled_evaluation_time)
    
    # Record the current simulation time as the time the car entered the fulfillment queue
    time_entered_queue_for_fulfillment = env.now

    # Store the time the car entered the fulfillment queue in the list_of_cars
    list_of_cars[car_id][14] = time_entered_queue_for_fulfillment

    # Request access to the fulfillment expert resource and yield the request
    with expert.request() as req:

        # Increment the fulfillment queue counter as the car enters the queue
        fulfillment_queue += 1

        yield req

        # Decrement the fulfillment queue counter as the car leaves the queue
        fulfillment_queue -= 1
    
        # Record the current simulation time as the time the car leaves the fulfillment queue
        time_left_queue_for_fulfillment = env.now
        
        # Store the time the car left the fulfillment queue in the list_of_cars
        list_of_cars[car_id][15] = time_left_queue_for_fulfillment
   
        # Calculate the time the car spent waiting in the fulfillment queue
        time_in_queue_for_fulfillment = time_left_queue_for_fulfillment - time_entered_queue_for_fulfillment

        # Store the time spent in the fulfillment queue in the list_of_cars
        list_of_cars[car_id][16] = time_in_queue_for_fulfillment
        
        # Generate a random sample for the fulfillment time based on a triangular distribution
        sampled_fulfillment_time = triangular_generator(min_fulfillment, max_fulfillment, mode_fulfillment)

        # Store the sampled fulfillment time in the list_of_cars
        list_of_cars[car_id][17] = sampled_fulfillment_time

        # Simulate the time it takes for the fulfillment to complete using env.timeout
        yield env.timeout(sampled_fulfillment_time)

    # Record the time the car left the system
    departure_time = env.now
    list_of_cars[car_id][18] = departure_time

    # Calculate the time spent in the system
    Time_Spent_in_System = list_of_cars[car_id][18] - list_of_cars[car_id][1]
    list_of_cars[car_id][19] = Time_Spent_in_System
```

```python
replication_number = 100  # Set the number of replications to run the simulation

list_of_data = []  # Initialize an empty list to store dataframes for each replication
list_of_measures = []  # Initialize an empty list to store dataframes for each replication's measures
used_seeds = set()  # Initialize an empty set to store used random seeds for ensuring uniqueness

for i in range(replication_number):

    # Generate a random seed for each replication
    seed = random.randint(0, 1000)

    # Ensure the generated seed is unique
    while seed in used_seeds:
        seed = random.randint(0, 1000)

    # Add the unique seed to the set of used seeds   
    used_seeds.add(seed)

    # Initialize the random number generator with the unique seed
    random.seed(seed)

    # Set initial queue counts to zero
    documentation_queue = 0  
    fulfillment_queue = 0  

    list_of_cars = []  # Initialize an empty list to store car information for each replication
    observation_time = []  # Initialize an empty list to store observation times for each replication

    # Initialize empty lists to store various queue lengths and resource usages during the simulation
    photo_queue_length = []  
    outside_queue_length = []  
    documentation_queue_length = []  
    fulfillment_queue_length = []  
    evaluation_queue_length = []  
    expert_queue_length = [] 

    photographer_usage = []  
    expert_usage = []  
    evaluation_usage = []  

    # Define given and calculated parameters for the simulation
    mean_arrival = 3.2 # Given (Changed after assessment)
    mean_photography = 6.0 # Given 
    min_documentation = 6.0 # Given (Changed after assessment)
    max_documentation = 10.0 # Given (Changed after assessment)
    mode_documentation = 8.0 # Given (Changed after assessment)
    mean_evaluation = 8.0 # Given (Changed after assessment)
    mean_complaint = 15.0 # Given
    min_fulfillment = 3.0 # Given (Changed after assessment)
    max_fulfillment = 4.0 # Given (Changed after assessment)
    mode_fulfillment = 3.5 # Given (Changed after assessment)

    # Set the number of resources for each process
    num_photographers = 2 # Given
    num_experts = 4 # Given (Changed after assessment)
    num_evaluation_experts = 3 # Given (Changed after assessment)
    observation_step = 2 # Arbitrary (Chosen in a way that serves both speed and accuracy)
    run_time_limit = 259200 # or six months!

    env = simpy.Environment () # Create a simulation environment

    # Create resources for each process
    photographer = simpy.Resource(env, capacity=num_photographers)
    expert = simpy.PriorityResource(env, capacity=num_experts)
    evaluation_expert = simpy.Resource(env, capacity=num_evaluation_experts)

    # Run the simulation processes using the car_generator and observe functions as separate processes
    env.process (car_generator(env, mean_arrival, mean_photography, mean_complaint,
                  min_documentation, max_documentation, mode_documentation,
                  mean_evaluation, min_fulfillment, max_fulfillment,
                  mode_fulfillment, photographer, expert, evaluation_expert))
                        
    env.process (observe(env, photographer, expert, evaluation_expert))

    env.run () # Run the simulation

    # Convert the list_of_cars to a pandas DataFrame
    data = pd.DataFrame(list_of_cars, columns = car_column_names)

    list_of_data.append(data) # Append the DataFrame to the list_of_data for each replication

    measure_data = pd.DataFrame([observation_time, photo_queue_length, outside_queue_length, documentation_queue_length, 
                                evaluation_queue_length, fulfillment_queue_length, expert_queue_length, photographer_usage, 
                                expert_usage, evaluation_usage]).T
    
    measure_data.columns = measure_column_names # Set the column names for the measure_data DataFrame

    list_of_measures.append(measure_data) # Append the DataFrame to the list_of_measures for each replication
```
