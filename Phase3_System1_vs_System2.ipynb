Now, before delving into the statistical tests for each metric, it is essential to determine the degree of freedom and calculate the standard error of the mean (SEM). These two factors play a crucial role in conducting reliable and accurate statistical analyses.

In the code block below, we have two functions defined: `DF_calculator` and `standard_error`.

1. **DF_calculator(var1, var2, r1, r2)**: This function calculates the degrees of freedom (df) for a statistical test involving two samples. It takes four input parameters: `var1`, `var2`, `r1`, and `r2`. `var1` and `var2` represent the variance of sample 1 and sample 2, respectively, while `r1` and `r2` are the sample sizes for the two samples. The function uses these parameters to compute the degrees of freedom (df) according to the formula:

  <img src="https://latex.codecogs.com/png.latex?\dpi{120}&space;\color{white}&space;df&space;=&space;\frac{(\frac{var_1}{r_1}&space;&plus;&space;\frac{var_2}{r_2})^2}{(\frac{(\frac{var_1}{r_1})^2}{r_1-1}&space;&plus;&space;\frac{(\frac{var_2}{r_2})^2}{r_2-1})}" alt="DF Formula">
</p>

   The calculated degrees of freedom (df) value is then returned by the function.

2. **standard_error(var1, var2, r1, r2)**: This function calculates the standard error of the mean (sem) for two samples. It also takes four input parameters: `var1`, `var2`, `r1`, and `r2`, representing the variance of sample 1, variance of sample 2, sample size of sample 1, and sample size of sample 2, respectively. The function computes the standard error of the mean using the formula:

<p align="center">
  <img src="https://latex.codecogs.com/png.latex?\dpi{120}&space;\color{white}&space;sem&space;=&space;\sqrt{\frac{var_1}{r_1}&space;&plus;&space;\frac{var_2}{r_2}}" alt="Standard Error Formula">
</p>

   The calculated standard error value is then returned by the function.

In summary, these functions provide a way to calculate degrees of freedom (df) and the standard error of the mean (sem) for statistical analyses involving two samples. They utilize the numpy library, which was conventionally imported as `np` for numerical computations, such as power and square root operations.


<span>

```python
# Define a function called 'DF_calculator' that takes four input parameters: 'var1', 'var2', 'r1', and 'r2'.
def DF_calculator(var1, var2, r1, r2):
    # Calculate the degrees of freedom (df) using the provided formula.
    # The formula involves variances of two samples and their respective sample sizes.
    df = np.power(((var1/r1) + (var2/r2)), 2) / (((np.power((var1/r1), 2)) / (r1-1)) + ((np.power((var1/r1), 2)) / (r2-1)))

    # Return the calculated degrees of freedom (df).
    return df

# Define another function called 'standard_error' that takes four input parameters: 'var1', 'var2', 'r1', and 'r2'.
def standard_error(var1, var2, r1, r2):
    # Calculate the standard error of the mean (sem) using the provided formula.
    # The formula involves variances of two samples and their respective sample sizes.
    sem = np.sqrt((var1/r1) + (var2/r2))

    # Return the calculated standard error of the mean (sem).
    return sem
```

<span style="font-family: Times New Roman; font-size: 13pt;">

In the next block, we have a series of calculations and comparisons related to two systems, system_1 and system_2, which are represented as DataFrames. The goal is to perform statistical comparisons between certain variables in both systems and determine if one system is better, indifferent, or if there is no significant difference.

1. `category = []`, `lower_bounds = []`, `upper_bounds = []`, and `comparison_list = []`: These lists will be used to store the names of the categories being compared, the lower bounds of confidence intervals, the upper bounds of confidence intervals, and the results of comparisons, respectively.

2. `for i in system_2.columns:`: This loop iterates through the columns of the DataFrame 'system_2'.

3. `category.append(i)`: The column name (i.e., category) is added to the 'category' list.

4. `ci = st.t.interval(...)`: This line calculates the confidence interval (ci) using Student's t-distribution. The `alpha=0.95` parameter specifies a 95% confidence level. The 'DF_calculator' function is used to calculate the degrees of freedom for the t-distribution. The `loc` parameter represents the difference between the average of 'system_1' and 'system_2' for the given category, and the `scale` parameter is the standard error of the mean difference calculated using the 'standard_error' function.

5. `lower = ci[0]`, `upper = ci[1]`: The lower and upper bounds of the confidence interval are extracted from 'ci'.

6. `lower_bounds.append(lower)`, `upper_bounds.append(upper)`: The lower and upper bounds are added to their respective lists.

7. The conditional statements inside the 'if' and 'else' blocks compare the confidence interval bounds to zero, and based on the results, the appropriate comparison (System 1 is better, System 2 is better, or Indifferent) is appended to the 'comparison_list'.

8. `if ('Probability' not in i) and ('Efficiency' not in i):`: This conditional statement checks if the column name (i.e., category) does not contain the words 'Probability' or 'Efficiency'. If it doesn't contain these words, the comparison is done differently based on whether the confidence interval contains zero or not.

9. `else:`: This 'else' block is executed when the column name contains the words 'Probability' or 'Efficiency', and the comparison is done differently based on the same condition (whether the confidence interval contains zero or not).

In summary, this code calculates confidence intervals for the difference between corresponding values in 'system_1' and 'system_2' DataFrames and performs statistical comparisons to determine which system is better or if they are indifferent. The results are stored in the 'comparison_list' for further analysis or reporting.

<span>

```python
# Initialize lists to store category names, lower bounds of confidence intervals, upper bounds of confidence intervals, and comparison results.
category = []
lower_bounds = []
upper_bounds = []
comparison_list = []

# Iterate through the columns of the 'system_2' DataFrame (because system 2 KPIs are a subset of system 1s)
for i in system_2.columns:

    # Store the name of the current category (column) in the 'category' list.
    category.append(i)

    # Calculate the confidence interval using Student's t-distribution for the current category.
    # "df" and "scale" arguments are calculated using functions defined earlier.
    ci = st.t.interval(alpha=0.95, df=DF_calculator(system_1.loc['Variance (S^2)'][i], system_2.loc['Variance (S^2)'][i], 10, 10),
                       loc=system_1.loc['Point Estimate (Average)'][i] - system_2.loc['Point Estimate (Average)'][i],
                       scale=standard_error(system_1.loc['Variance (S^2)'][i], system_2.loc['Variance (S^2)'][i], len(system_1)-4, len(system_2)-4))

    # Extract the lower and upper bounds of the confidence interval and store them in 'lower' and 'upper' variables.
    lower = ci[0]
    upper = ci[1]

    # Add the lower and upper bounds to the 'lower_bounds' and 'upper_bounds' lists, respectively.
    lower_bounds.append(lower)
    upper_bounds.append(upper)

    # Compare the confidence interval bounds to zero and append the appropriate result to the 'comparison_list' based on conditions.
    # The higher the probabilty and the efficiency, the better.
    # The lower the queue length and queuing, the better.
    # and since we have subtracted system 1 metrics from system 2s, if the entire confidence interval is above 0, system 1 has a better performance on probabilities and efficiency
    # but if the entire confidence interval is below 0, system 1 has a worse performance on queues

    if ('Probability' not in i) and ('Efficiency' not in i):
        # If the confidence interval straddles zero, we consider the systems as "Indifferent."
        if (0 > lower) and (0 < upper):
            comparison_list.append("Indifferent")
        # If the confidence interval is entirely below zero, we consider "System 1" as better.
        elif (0 > lower) and (0 > upper):
            comparison_list.append("System 1 Is Better")
        # If the confidence interval is entirely above zero, we consider "System 2" as better.
        elif (0 < lower) and (0 < upper):
            comparison_list.append("System 2 Is Better")
    else:
        # The same comparisons are made as above, but in this case, the words "System 1" and "System 2" are switched, because here, the higher the better as explained earlier. 
        if (0 > lower) and (0 < upper):
            comparison_list.append("Indifferent")
        elif (0 > lower) and (0 > upper):
            comparison_list.append("System 2 Is Better")
        elif (0 < lower) and (0 < upper):
            comparison_list.append("System 1 Is Better")
```

<span style="font-family: Times New Roman; font-size: 13pt;">

In the following code, `comparison_list`, which contains the results of statistical comparisons between two systems for different categories, is converted into a DataFrame called `comparison_data` with comparison results stored as columns. Then, the DataFrame is transposed to have categories as rows and different systems as columns, resulting in `comparison_transposed`. This operation allows for a more convenient representation of the comparison results, with each row corresponding to a specific category, and the columns indicating the results for the two systems being compared.

<span>

```python
# Convert the list of comparison results to a DataFrame with categories as columns and a single row.
comparison_data = pd.DataFrame(comparison_list).T

# Set the column names of the DataFrame to match the columns of 'system_2' for proper labeling.
comparison_data.columns = list(system_2.columns)

# Transpose the DataFrame to have categories as rows and different systems as columns.
comparison_transposed = comparison_data.T
```

<span style="font-family: Times New Roman; font-size: 13pt;">

In the following code block, a plot is created to visualize the statistical comparisons made earlier. The following steps are taken to construct the plot:

1. The font family for the plot is set to Times New Roman using `plt.rcParams['font.family'] = 'Times New Roman'`.

2. The number of intervals (categories) is calculated, along with the number of columns and rows required to arrange the subplots in a grid.

3. A subplot grid with the specified number of rows and columns is created, and the axes are flattened into a 1D array for easier iteration.

4. The code iterates through each interval, and for each one, it retrieves the lower and upper bounds of the confidence interval and determines which color to use based on the comparison result.

5. A subplot is selected, and a horizontal line (axhline) is drawn at y=-0.02 as a reference line. A plot is created with markers at the start and end points of the confidence interval, representing the range of values. The color used for the plot depends on the comparison result for that interval.

6. The axis labels and title for each subplot are set, and the x-axis limits are adjusted to accommodate the data's range.

7. If there are remaining empty subplots (when the number of intervals is not a multiple of the number of columns), they are hidden.

8. A custom legend is created to explain the color codes used in the plot. It includes two colored patches (rectangles) representing the colors used for "System 1 Is Better" and "System 2 Is Better."

9. The spacing between subplots is adjusted using `plt.tight_layout()` to prevent overlapping.

10. The subplot is moved down slightly using `plt.subplots_adjust(top=0.96)` to make space for the legend.

11. The plot is displayed using `plt.show()`.

In summary, this code generates a plot that visually represents the statistical comparisons between `system_1` and `system_2` for different categories. Each category's confidence interval is plotted as a line with markers, and the color of the line depends on whether "System 1 Is Better," "System 2 Is Better," or "Indifferent." The plot includes a custom legend to explain the color scheme used in the visualization.

By using this plot, we can visually compare these two systems on all our 28 metrics and gain valuable insights into how each of them performed.

<span>

```python
# Set the font family to Times New Roman
plt.rcParams['font.family'] = 'Times New Roman'

# Calculate the number of rows and columns for the subplot grid
num_intervals = len(category)
num_cols = 2
num_rows = (num_intervals + num_cols - 1) // num_cols

# Create the subplot grid
fig, axes = plt.subplots(num_rows, num_cols, figsize=(17, 24))

# Flatten the axes array to iterate through the subplots
axes = axes.flatten()

# Iterate through each interval and plot it in the corresponding subplot
for i in range(num_intervals):
    # Define the interval (start and end points) from the lower and upper bounds of the confidence interval
    start_point = lower_bounds[i]
    end_point = upper_bounds[i]

    # Define colors to be used in the plot
    colors = ['#33FF57', '#FF5733', '#33A3FF']

    # Determine the color to be used based on the comparison result for the current interval
    if comparison_data[category[i]][0] == 'System 1 Is Better':
        color_picker = 0
    elif comparison_data[category[i]][0] == 'System 2 Is Better':
        color_picker = 1
    elif comparison_data[category[i]][0] == 'Indifferent':
        color_picker = 2

    # Get the corresponding subplot for this interval
    ax = axes[i]

    # Create the plot in the subplot using markers to represent the confidence interval
    plot = ax.plot([start_point, end_point], [0, 0], marker='o', markersize=10, color=colors[color_picker])

    # Add a horizontal reference line at y=-0.02 for better visualization
    ax.axhline(y=-0.02, color='gray', linestyle='--')

    # Set axis labels and title for the subplot
    ax.set_yticks([])  # Hide the y-axis ticks
    ax.set_title(category[i], fontsize=12)

    # Set the x-axis limits for the subplot, adding some padding to accommodate the data range
    interval = max(abs(lower_bounds[i]), abs(upper_bounds[i]))
    ax.set_xlim(0 - (interval * 1.2), 0 + (interval * 1.2))

# Hide any remaining empty subplots
for i in range(num_intervals, num_rows * num_cols):
    axes[i].axis('off')

# Create legend handles and labels for the custom legend
legend_handles = [mpatches.Patch(color=colors[0]), mpatches.Patch(color=colors[1]), mpatches.Patch(color=colors[2])]
legend_labels = ['System 1 Is Better', 'System 2 Is Better', 'Indifferent']

# Add a custom legend for the three colors to explain the comparison results
fig.legend(handles=legend_handles, labels=legend_labels, loc='upper center', ncol=3, fontsize=15)

# Adjust spacing between subplots and show the plot
plt.tight_layout()

# Move the subplot down to make space for the legend
plt.subplots_adjust(top=0.96)

# Show the plot
plt.show()
```

in the next code block, the `attributes` list is created to store the column names of the DataFrame `comparison_data`, which represent the different attributes or categories being compared between `system_1` and `system_2`. Two new lists, `system1_check` and `system2_check`, are initialized to store checkmark ('\u2713') or dash ('-') symbols based on the comparison results.

The code then iterates through the rows of `comparison_transposed`, which is the transposed DataFrame of `comparison_data`. For each row (i.e., for each attribute/category), the code checks the corresponding value in the first column (index 0) of `comparison_transposed`.

- If the comparison result for the attribute/category is "System 1 Is Better," a checkmark symbol is appended to `system1_check` to indicate that System 1 is better, and a dash symbol is appended to `system2_check` to indicate that System 2 is not better.

- If the comparison result is "System 2 Is Better," a checkmark symbol is appended to `system2_check` to indicate that System 2 is better, and a dash symbol is appended to `system1_check` to indicate that System 1 is not better.

- If the comparison result is "Indifferent" (meaning there is no significant difference between the systems for that attribute/category), dash symbols are appended to both `system1_check` and `system2_check`.

As a result, after this code block execution, `system1_check` and `system2_check` will contain checkmark symbols or dashes to represent the comparison results for each attribute/category between `system_1` and `system_2`, making it easier to visualize which system is better for each attribute.

<span>

```python
# Create a list 'attributes' to store the column names of the DataFrame 'comparison_data',
# representing the different attributes or categories being compared between 'system_1' and 'system_2'.
attributes = list(comparison_data.columns)

# Initialize two lists, 'system1_check' and 'system2_check', to store checkmark ('\u2713') or dash ('-') symbols based on the comparison results.
system1_check = []
system2_check = []

# Iterate through the rows of the transposed DataFrame 'comparison_transposed',
# where each row represents a specific attribute/category.
for i in range(len(comparison_transposed[0])):

    # Check the corresponding value in the first column (index 0) of 'comparison_transposed',
    # which represents the comparison result for the current attribute/category.
    if comparison_transposed[0][i] == "System 1 Is Better":
        # If the comparison result is "System 1 Is Better," append a checkmark symbol to 'system1_check'
        # to indicate that System 1 is better, and append a dash symbol to 'system2_check'
        # to indicate that System 2 is not better for this attribute/category.
        system1_check.append('\u2713')
        system2_check.append('-')

    elif comparison_transposed[0][i] == "System 2 Is Better":
        # If the comparison result is "System 2 Is Better," append a checkmark symbol to 'system2_check'
        # to indicate that System 2 is better, and append a dash symbol to 'system1_check'
        # to indicate that System 1 is not better for this attribute/category.
        system1_check.append('-')
        system2_check.append('\u2713')

    else:
        # If the comparison result is "Indifferent" (no significant difference between the systems),
        # append dash symbols to both 'system1_check' and 'system2_check' to indicate that
        # neither system is better for this attribute/category.
        system1_check.append('-')
        system2_check.append('-')
```

<span style="font-family: Times New Roman; font-size: 13pt;">

Finally, a DataFrame named `comparison` is created to summarize the results of the comparison between `system_1` and `system_2` for various attributes. The DataFrame combines three lists: `attributes` (attribute names), `system1_check` (checkmarks if `system_1` is better), and `system2_check` (checkmarks if `system_2` is better). It transposes the DataFrame to have attributes as rows and columns representing the attribute name, _System 1 Is Better_,  and _System 2 Is Better_. Afterward, the code calculates the counts of attributes for which `system_1` is better (`count_system1`) and `system_2` is better (`count_system2`). A new row labeled _Overall Assessment_ is created with these counts and appended to the `comparison` DataFrame. The DataFrame is then exported to an Excel file named `System 1 vs System 2.xlsx` for further analysis or reporting.

<span>

```python
# Create a DataFrame 'comparison' to summarize comparison results for attributes, 'System 1 Is Better', and 'System 2 Is Better'.
comparison = pd.DataFrame([attributes, system1_check, system2_check]).T
comparison.columns = ['Factor', 'System 1 Is Better', 'System 2 Is Better']  # Set column names for clarity.

# Calculate counts of attributes where 'system_1' is better ('count_system1') and 'system_2' is better ('count_system2').
count_system1 = comparison.loc[comparison['System 1 Is Better'] == '\u2713'].count()[0]
count_system2 = comparison.loc[comparison['System 2 Is Better'] == '\u2713'].count()[0]

# Create an 'Overall Assessment' row and append it to the 'comparison' DataFrame.
new_row = pd.DataFrame([['Overall Assessment', count_system1, count_system2]], columns=comparison.columns)
comparison = pd.concat([comparison, new_row], ignore_index=True)

# Export 'comparison' DataFrame to "Phase 3 - Comparison.xlsx" for further analysis or reporting.
comparison.to_excel("E:\Desktop\Phase 3 - Comparison.xlsx")
```
